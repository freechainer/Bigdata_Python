```python
def createData(yy, mm):
        fp="files/kto_{}{}.xlsx".format(yy,mm)
        df=pd.read_excel(fp, header=1, skipfooter=4, usecols='A:G')
        df['기준년월']='{}-{}'.format(yy,mm)
        continents_list=['아시아주','미주','구주','대양주','아프리카주','기타대륙','교포소계']
        cond=(df.국적.isin(continents_list)==False)
        dfCountry=df[cond].reset_index(drop=True)
        continents = ['아시아']*25+['아메리카']*5+['유럽']*23+['오세아니아']*3+['아프리카']*2+['기타대륙']+['교포']
        dfCountry['대륙']=continents
        dfCountry['관광객비율(%)']=round(dfCountry['관광']/dfCountry['계']*100,1)
        dfCountry['전체비율(%)']=round(dfCountry['관광']/ts*100 ,2)
        return(dfCountry)
        
#         print(df)
#         print("="*50)
```

```python
df=pd.DataFrame()
for yy in range(2010,2020):
    for mm in range(1,13):
                try:
                    temp=createData(str(yy),str(mm).zfill(2))
                    df=df.append(temp, ignore_index=True)
                except: #예외가 발생되면 (파일이 없다면)
                    pass #건너뛰어라
```
