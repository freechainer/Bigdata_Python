```python
"""
1. []

[]사이의 문자와 매치
정규식 : [A]
문자열 : A(매치됨), B(매치안됨)
정규식 : [ABCDE]
문자열 : A(매치됨), B(매치됨), ABC(매치됨)

2.\d

[0-9] 0부터 9까지의 숫자를 의미함

3. \d

[^0-9] 0부터 9까지의 숫자를 제외한 모든 문자

4. *

*앞에 등장한 문자에 대해 0개 이상의 문자

정규표현식 : ca*t
문자열 : ct(매치), cat(매치), caaaaaat(매치)

5. +

+앞에 등장한 문자에 대해 1개 이상의 문자

정규표현식 ca*t
문자열 : ct(매치X), cat(매치), caaaaaaaat(매치), caaaaak(매치X)

6. {}:반복

정규표현식 : ca{3}t
c다음에 a가 3개 나오고 t가 이어서 나와야 함
문자열 : caat(매치X), caaat(매치)

{2,5}: 2번이상 5번이하
정규표현식 : ca{2,5}t
문자열 : cat(매치X), caat(매치), caaat(매치)

7. ? : 있어도 없어도 됨

ca?t : a는 있어도 없어도 가능하다
ct(매치), cat(매치)

"""
```
