```python
# 표준화, 정규화

# 표준화 = (내점수 - 평균) / 표준편차 => 1                  -> 표준편차에 대해서 1배 더 받음
# 표준화 = (내점수 - 평균) / 표준편차 => 2                  -> 표준편차에 대해서 2배 더 받음
# 표준화 = (내점수 - 평균) / 표준편차 => -5               -> 표준편차에 대해서 -5점이 작다

# 정규화 = (각 데이터-평균) / (최대값-최소값)

def myZscore(x):
    return (x-x.mean())/x.std()
```

```python
z=df.groupby('year').lifeExp.transform(myZscore)
z
```
